### Дано:

```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}
```

### Программа-секретарь:

```python
def user_number_doc(): #получает пользовательские данные о № документа
    user_input = input('Введите номер документа: ')
    return user_input

def code_number_doc(): #получает данные о номерах документов из documents
    doc = []
    num = 0
    for numbers in documents:
        docs = (list(documents[num].values()))
        doc.append(docs[1])
        num += 1
    return doc

def code_places(): #получает словари формата "номер документа : полка"
    dict_shelf = {}
    num = len(documents) -1
    shelf_list = (list(directories.items()))
    for shelf_one in shelf_list:
        num2 = len(shelf_one[1]) -1
        for element in shelf_one[1]:
            dict_shelf[shelf_one[1][num2]] = shelf_one[0]
            num2 -=1
    return dict_shelf
 
def shelf(): #получает текущие данные о номерах полок
    text_shelf = list(directories.keys())
    return text_shelf
        
def user_places(): #получает информацию о полке по номеру документа от пользователя
    text_out = 'Документ не найден в базе'
    user_input = user_number_doc()
    shelf_list = (list(directories.items()))
    for shelf in shelf_list:
        if shelf[1].count(user_input) == 1:
            text_out = 'Документ хранится на полке: ' + shelf[0]
    return text_out

def info(): #получает полную информацию о документах
    num = 0
    list_out = []
    text = ''
    text_out = ''
    for element in code_number_doc():
        val = (list(documents[num].values()))
        num += 1
        list_out.append(val)
    for element2 in list_out:
        element2.append(code_places().get(element2[1]))
        text = f"№: {element2[1]}, тип: {element2[0]}, владелец: {element2[2]}, полка хранения: {element2[3]}\n"
        text_out = text_out + text
    return text_out

def user_shelf(): #получает пользователские данные о номере полки
    user_input = input('Введите номер полки: ')
    return user_input

def add_shelf(): #добавляет новую полку
    new_shelf = '0'
    shelf_add = shelf()
    user_input = user_shelf()
    ads = ""
    if shelf_add.count(user_input) == 0 and user_input != '0':
        shelf_add.append(user_input)
        directories[user_input] = []
        new_shelf = str(user_input)
        ads = f"Полка {user_input} добавлена"
    elif shelf_add.count(user_input) != 0:
        ads = f"Полка с номером {user_input} уже существует"
    elif user_input == '0':
        ads = f"Нельзя создать полку с номером 0"
    return ads

def owner(): #находит владельца документа по номеру от пользователя
    text_out = '0'
    user_input = user_number_doc()
    docs_len = len(documents) -1    
    for string in documents:
        if documents[docs_len].get('number') == user_input:
            text_out = 'Владелец документа: ' + documents[docs_len].get('name')
        else:
            text_out = 'Документ не найден в базе'
            docs_len -=1
    return text_out

def command_line(): #получает пользовательские данные о команде
    user_input = input('Введите команду: ')
    return user_input    

def del_shelf(): #удаляет пустую полку
    user_input = user_shelf()
    active_shelfs = list(code_places().values())
    no_shelf_try = shelf()
    
    if active_shelfs.count(user_input) != 0:
        del_sh = 'На полке есть документы, удалите их перед удалением полки'
    elif no_shelf_try.count(user_input) == 0:
        del_sh = 'Такой полки не существует'
    else:
        del(directories[user_input])
        del_sh = 'Полка удалена'
    return del_sh

def user_add_doc(): #добавление документа в базу
    add_doc = ''
    user_num = user_number_doc()
    num = code_number_doc()
    user_type = input('Введите тип документа: ')
    user_owner = input('Введите владельца документа: ')
    code_sh = shelf()
    user_sh = user_shelf()
    if num.count(user_num) != 0:
        add_doc = 'Документ с таким номером уже есть в базе'
    elif code_sh.count(user_sh) == 0:
        add_doc = 'Такой полки не существует'
    else:
        new_doc_dict = {}
        new_doc_dict['type'] = user_type
        new_doc_dict['number'] = user_num
        new_doc_dict['name'] = user_owner
        documents.append(new_doc_dict)
        user_list = []
        user_list.append(user_num)
        directories[user_sh] = user_list
        add_doc = 'Ввод документа успешно завершен'
    return add_doc
        
def user_del_doc(): #удаление документа из базы
    user_input = user_number_doc()
    num = code_number_doc()
    if num.count(user_input) == 0:
        del_doc = 'Документа с таким номером нет в базе'
    else:
        for element in documents:
            if element.get('number') == user_input:
                for district in directories:
                    element2 = (directories.get(district))
                    if element2.count(user_input) != 0:
                        element2.remove(user_input)
                documents.remove(element)
        del_doc = 'Удаление документа успешно завершено'
    return del_doc

def user_change_place(): #перенос документа
    user_num = user_number_doc()  
    user_sh = user_shelf()
    num = code_number_doc()
    code_sh = shelf()
    try_same_sh = (code_places())

    if num.count(user_num) == 0:
        change_pl = 'Документа с таким номером нет в базе'    
    elif code_sh.count(user_sh) == 0:
        change_pl = 'Такой полки не существует'
    elif try_same_sh.get(user_num) == user_sh:
        change_pl = 'Документ уже находится на этой полке'
    else:
        for district in directories:
            element2 = (directories.get(district))
            if element2.count(user_num) != 0:
                element2.remove(user_num)
            if district == user_sh:
                element2.append(user_num)
        change_pl = 'Перенос документа завершен'

def main():
    while True:
        user_input = command_line()
        if user_input == 'p': #находит владельца документа по номеру от пользователя
            print(owner()) 
        elif user_input == 's': #получает информацию о полке по номеру документа от пользователя
            print(user_places())
        elif user_input == 'l': #получает полную информацию о документах
            print(info()) 
        elif user_input == 'ads': #добавляет новую полку
            print(add_shelf())
            print('Текущие номера полок: ', shelf())
        elif user_input == 'ds': #удаляет пустую полку
            print(del_shelf())
            print('Текущие номера полок: ', shelf())
        elif user_input == 'ad': #пользователь добавляет новый документ
            print(user_add_doc())
            print('В базе хранятся следующие документы:')
            print(info())            
        elif user_input == 'd': #удаление документа из базы
            print(user_del_doc())
            print('В базе хранятся следующие документы:')
            print(info())            
        elif user_input == 'm': #перенос документа
            print(user_change_place())
            print('В базе хранятся следующие документы:')
            print(info())       
        elif user_input == 'q': #выход
            print('Конец программы') 
            break


main()
```
